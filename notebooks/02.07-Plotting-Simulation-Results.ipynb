# Plotting Simulation Results



```python
from archetypal import IDF

idf = IDF(
    "data/seed_file_V960.idf",
    epw="data/USA_CA_Chino.AP.722899_TMYx.epw",
    readvars=False,  # readvars=False, will make the simulation complete faster 
    # (skipping the reading of the eplusout.eso file)
)
```

```python
idf.simulate()
```

# Plotting OutputMeters and OutputVariables



```python
ds_heating = idf.meters.OutputMeter.DistrictHeating__Facility.values("kWh")
```

```python
ds_heating.plot()
```

## Producing the heating load duration curve

```python
ds_heating.sort_values(ignore_index=True, ascending=False).plot()
```

## Producing 2D plots

```python
fig, ax = ds_heating.plot2d(vmin=0, cmap="Reds", figsize=(10, 5), dpi=144)
```

## Plotting a 3D plot

```python
fig, ax = ds_heating.plot3d(axis_off=False, figsize=(10, 5), dpi=144)
```
