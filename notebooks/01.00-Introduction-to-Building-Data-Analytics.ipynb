# Introduction to Building Data Analytics

In this chapter, we are going to take what we learned about design spaces in the previous chapter, and apply it for the study of design analytics for single building projects. We will go over generating 3 types of deliverables:

1. The Energy End Use Characterization
2. The Envelope Sensitivity Analysis
3. The Parametric Analysis

## The Energy End Use Characterization

The purpose of this analysis is to visualize the key drivers of a buildings annual energy use for a specific design. By presenting a holistic view of the energy sources and sinks in the building, it can indicate which elements of the building draw the most energy and potentially target design modifications of those elements.

The deliverable from this analysis is a Sankey diagram of energy flows in the building. The process below outlines the steps needed to create this diagram from an EnergyPlus model.

## Envelope Sensitivity Analysis

The purpose of this analysis is to understand the sensitivity of the buildings TEDI (thermal energy demand intensity), CEDI (cooling energy demand intensity)  and, optionally, other performance metrics such as EUI (energy use intensity), peak cooling demand, peak heating demand, etc.  on envelope properties. In this analysis, the basis of design or baseline design energy model serves as the reference, and its major envelope properties (e.g., WWR, wall R-value, roof R-value, infiltration rate, etc.) are varied one at a time.

The deliverable from the analysis is a set of line plots with the dependent variable (TEDI, CEDI, etc.) on the y-axis, and the parameter value on the x-axis (Figure 2). The line represents the overall trend for each parameter. The baseline values are marked by dark blue dots, and the recommended value is marked with a light blue circle.

## Parametric Analysis

The purpose of this analysis is to visualize the effect of different combinations of building parameters (e.g., wall R-value, window U-value, etc.) on its EUI or other performance metrics. This can allow a designer to see the impacts of constraining one or more parameters to a specific value or range of values on the building's energy use, or to determine all possible parameter combinations that result in the target energy use values.

The deliverable is a parallel coordinates plot, with one vertical axis for each parameter (Figure 3). The last axis (or several axes) shows the EUI (or several performance metrics) resulting from the selected parameter combinations. The dark lines represent parameter combinations simulated using the whole building energy model. The lighter lines represent results from all other combinations that are generated using interpolation between simulated values.
